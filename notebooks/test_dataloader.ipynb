```python
import matplotlib.pyplot as plt
from data.datasets import CityscapesDataset
from data.preprocessing import BasicTransform
```

```python
# Initialize transforms
transform = BasicTransform()

# Test clear dataset
clear_dataset = CityscapesDataset(mode='train', foggy=False, transforms=transform)
img, annotation = clear_dataset[0]
print("Clear Image Shape:", img.shape, "Annotation Shape:", annotation.shape)
```

```python
# Visualization for clear dataset
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].imshow(img.permute(1, 2, 0))
axes[0].set_title('Clear Image')
axes[0].axis('off')

axes[1].imshow(annotation, cmap='gray')
axes[1].set_title('Annotation')
axes[1].axis('off')

plt.show()
```

```python
# Visualization for foggy dataset
# Test foggy dataset again after ensuring files are local
foggy_dataset = CityscapesDataset(mode='train', foggy=True, transforms=transform)
foggy_img = foggy_dataset[0]
print("Foggy Image Shape:", foggy_img.shape)

# Visualization
plt.figure(figsize=(6, 5))
plt.imshow(foggy_img.permute(1, 2, 0))
plt.title('Foggy Image')
plt.axis('off')
plt.show()
```
